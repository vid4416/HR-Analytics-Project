```python
#PREPROCESSING THE DATA
#In our dataset, we see that there are 4 unwanted columns
#They are EmployeeCount, Over18, EmployeeNumber and StandardHour. 
#So, we have to remove these columns for more accuracy. 
```

```python
df.drop('EmployeeCount', axis = 1, inplace = True)
df.drop('StandardHours', axis = 1, inplace = True)
df.drop('EmployeeNumber', axis = 1, inplace = True)
df.drop('Over18', axis = 1, inplace = True)
print(df.shape)
```

```python
#We have successfully deleted the unwanted columns
```

```python
y = df.iloc[:, 1]
x = df
x.drop('Attrition', axis = 1, inplace = True)
```

```python
#Label Encoding
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
```

```python
#In our dataset, there are 7 categorical data which should be converted to numerical(int). 
#Hence, we have to create 7 dummy variales
```

```python
dummies_BusinessTravel = pd.get_dummies(df['BusinessTravel'],
                                    prefix ='BusinessTravel')
dummies_Department = pd.get_dummies(df['Department'],
                                prefix ='Department')
dummies_EducationField = pd.get_dummies(df['EducationField'],
                                    prefix ='EducationField')
dummies_Gender = pd.get_dummies(df['Gender'],
                            prefix ='Gender', drop_first = True)
dummies_JobRole = pd.get_dummies(df['JobRole'],
                             prefix ='JobRole')
dummies_MaritalStatus = pd.get_dummies(df['MaritalStatus'],
                                   prefix ='MaritalStatus')
dummies_OverTime = pd.get_dummies(df['OverTime'],
                              prefix ='OverTime', drop_first = True)
```

```python
# Adding these dummy variables to input X
X = pd.concat([x, dummies_BusinessTravel, dummies_Department,
               dummies_EducationField, dummies_Gender, dummies_JobRole,
               dummies_MaritalStatus, dummies_OverTime], axis = 1)
# Deleting the categorical data
X.drop(['BusinessTravel', 'Department', 'EducationField',
        'Gender', 'JobRole', 'MaritalStatus', 'OverTime'],
        axis = 1, inplace = True)
 
print(X.shape)
print(y.shape)
```

```python
#Splitting data using train test split
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size = 0.25, random_state = 40)
```

```python
#We have successfully completed preprocessing steps. 
#Now, we will use KNeighbors Classifier to find best neighbor 
```

```python
from sklearn.neighbors import KNeighborsClassifier
neighbors = []
cv_scores = []
```

```python
#Using 10 fold cross validation score
```

```python
from sklearn.model_selection import cross_val_score
for k in range(1, 40, 2):
    neighbors.append(k)
    knn = KNeighborsClassifier(n_neighbors = k)
    scores = cross_val_score(
        knn, X_train, y_train, cv = 10, scoring = 'accuracy')
    cv_scores.append(scores.mean())
error_rate = [1-x for x in cv_scores]
```

```python
import warnings
warnings.simplefilter("ignore")
```

```python
# determining the best k
optimal_k = neighbors[error_rate.index(min(error_rate))]
print('The optimal number of neighbors is % d ' % optimal_k)
   
# plotting between misclassification error and k
plt.figure(figsize = (10, 6))
plt.plot(range(1, 40, 2), error_rate, color ='blue', linestyle ='dashed', marker ='o',
         markerfacecolor ='red', markersize = 10)
plt.xlabel('Number of neighbors')
plt.ylabel('Misclassification Error')
plt.show()
```

```python
#Prediction Score 
```

```python
from sklearn.model_selection import cross_val_predict, cross_val_score
from sklearn.metrics import accuracy_score, classification_report
from sklearn.metrics import confusion_matrix
```

```python
def print_score(clf, X_train, y_train, X_test, y_test, train = True):
    if train:
        print("Train Result:")
        print("------------")
        print("Classification Report: \n {}\n".format(classification_report(
                y_train, clf.predict(X_train))))
        print("Confusion Matrix: \n {}\n".format(confusion_matrix(
                y_train, clf.predict(X_train))))
 
        res = cross_val_score(clf, X_train, y_train,
                              cv = 10, scoring ='accuracy')
        print("Average Accuracy: \t {0:.4f}".format(np.mean(res)))
        print("Accuracy SD: \t\t {0:.4f}".format(np.std(res)))
        print("accuracy score: {0:.4f}\n".format(accuracy_score(
                y_train, clf.predict(X_train))))
        print("----------------------------------------------------------")
    elif train == False:
        print("Test Result:")
        print("-----------")
        print("Classification Report: \n {}\n".format(
                classification_report(y_test, clf.predict(X_test))))
        print("Confusion Matrix: \n {}\n".format(
                confusion_matrix(y_test, clf.predict(X_test))))
        print("accuracy score: {0:.4f}\n".format(
                accuracy_score(y_test, clf.predict(X_test))))
        print("-----------------------------------------------------------")
         
knn = KNeighborsClassifier(n_neighbors = 7)
knn.fit(X_train, y_train)
print_score(knn, X_train, y_train, X_test, y_test, train = True)
print_score(knn, X_train, y_train, X_test, y_test, train = False)
```
